```python
# load the dataset
import pandas as pd
import numpy as np
df = pd.read_csv('/content/dirty_cafe_sales.csv')
df.head()
```

The dataset is of a cafe sales with 8 coloumns

```python
df.tail()
```

```python
df.isnull().sum()
```

```python
df.info()
```

info() gives the over view of whole data set each coloumn consiste of none null values but as we seen before first five rows of data set the most values of coloumns were filled with "ERRORS" and "UNKNOWN".As we look closely data types of coloumns were mismatched

```python
# to change the data types of coloumns
df['Quantity'] = pd.to_numeric(df['Quantity'],errors = 'coerce')
df['Total Spent'] = pd.to_numeric(df['Total Spent'],errors = 'coerce')
df['Price Per Unit'] = pd.to_numeric(df['Price Per Unit'],errors = 'coerce')
df['Total Spent'] = pd.to_numeric(df['Total Spent'],errors = 'coerce')
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'],errors = 'coerce')
df.info() # again info is used to check if data types have changed
```

Now the datatypes have been changed by using pandas library now it's time to remove null values

```python
# Replace error and unknown values with nan
df = df.replace({'ERRORS': np.nan, 'UNKNOWN': np.nan, 'ERROR': np.nan})
df.head()
```

```python
# fill those null values
df['Item']=df['Item'].fillna('UNKNOWN')
df['Quantity'] = df['Quantity'].fillna(df['Total Spent'] / df['Price Per Unit'])
df['Total Spent'] = df['Total Spent'].fillna(df['Quantity'] * df['Price Per Unit'])
df['Payment Method'] = df['Payment Method'].fillna('UNKNOWN')
df['Location'] = df['Location'].fillna('UNKNOWN')
df['Transaction Date'] = df['Transaction Date'].fillna('UNKNOWN')
df['Price Per Unit']=df['Price Per Unit'].fillna(df['Quantity'] / df['Total Spent'])
df.dropna(subset=['Quantity','Price Per Unit','Total Spent'], inplace = True)
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.to_csv('cleaned_cafedata')
```
